## Descriptive Statistics

```python
#Using updated traffic data by NDN Surya 
df_updated_data =pd.read_csv('updated_data.csv')
df_updated_data.head()
```

### Column: type

```python
df_updated_data['type'].value_counts()
```

The descriptive statistics for this column: **type** have been skipped as it contains only one unique value.

### Column: traffic_level

```python

data_desc_traffic_level = df_updated_data['traffic_level'].describe()
data_desc_traffic_level
```

```python
print("Count for traffic_level: ", data_desc_traffic_level['count'])
print("Mean for traffic_level: ", data_desc_traffic_level['mean'])
print("Min value for traffic_level: ", data_desc_traffic_level['min'])
print("Max value for traffic_level: ", data_desc_traffic_level['max'])

#Median
median_traffic_level = df_updated_data['traffic_level'].median()
print("Median for traffic_level: ", median_traffic_level)

#Mode
mode_traffic_level = df_updated_data['traffic_level'].mode()
print("Mode for traffic_level: ", mode_traffic_level)

#Range
range_value_traffic_level = data_desc_traffic_level['max'] - data_desc_traffic_level['min']
print("Range for traffic_level:", range_value_traffic_level)

#Variance
variance_value_traffic_level = df_updated_data['traffic_level'].var()
print("Variance for traffic_level:", variance_value_traffic_level)

#Standard Deviation
standard_deviation_traffic_level = df_updated_data['traffic_level'].std()
print("Standard Deviation for traffic_level:", standard_deviation_traffic_level)

#IQR
q1_traffic_level = df_updated_data['traffic_level'].quantile(0.25) #25th percentile
q3_traffic_level = df_updated_data['traffic_level'].quantile(0.75)  # 75th percentile
iqr_traffic_level = q3_traffic_level - q1_traffic_level
print("IQR for traffic_levl:", iqr_traffic_level)
```

#### Visualization

```python
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
#Box Plot for min, max, median, and IQR
sns.boxplot(df_updated_data['traffic_level'])
plt.title("Box Plot showing min, max, median, and IQR")
plt.show()
```

```python
#Bar plot for the central tendency values and variance
#Dictionary of statistics
stats_traffic_level = {
    "Mean" : data_desc_traffic_level['mean'],
    "Median": median_traffic_level,
    "Mode": mode_traffic_level[0],
    "Min": data_desc_traffic_level['min'],
    "Max": data_desc_traffic_level['max'],
    "Range": range_value_traffic_level,
    "Variance": variance_value_traffic_level
}

#Plotting
plt.bar(stats_traffic_level.keys(), stats_traffic_level.values(), color=['blue', 'red', 'green', 'orange','purple', 'black', 'aqua'], width=0.4)
plt.title("Bar plot for the central tendency values and variance of traffic_level")
plt.ylabel("traffic_level")
plt.show()
```

```python
#Plotting standard deviation
plt.bar('Data', data_desc_traffic_level['mean'], yerr=standard_deviation_traffic_level, capsize=5, color='lightblue')
```

### Column: coordinates

Column: **coordinates** were excluded from the descriptive statistical analysis as it represents geospatial data, which do not follow the same statistical properties as numerical or categorical variables. Descriptive statistics such as mean, median, and variance are used to summarize distributions of numeric or categorical data, but they are not appropriate for geospatial coordinates. 

### Column: date

```python
traffic_level_group_by_date= df_updated_data.groupby(df_updated_data['date'])['traffic_level'].mean()
traffic_level_group_by_date.head()
```

#### Visualization

```python
import matplotlib.pyplot as plt
```

```python
#Plotting the data using Line Plot
plt.figure(figsize=(16,6))
plt.plot(traffic_level_group_by_date.index,traffic_level_group_by_date.values, marker='o', color='b', linestyle='-', linewidth=2)
plt.title("Average Traffic Level Grouped by Date")
plt.xlabel("Date")
plt.ylabel("Average Traffic Level")
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#Plotting the data using Bar Plot
plt.figure(figsize=(16, 6))
traffic_level_group_by_date.plot(kind='bar', color='skyblue')
plt.title("Average Traffic Level by Date")
plt.xlabel("Date")
plt.ylabel("Average Traffic Level")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Column: time

```python

df_updated_data['hour'] = pd.to_datetime(df_updated_data['time'], format='%H:%M:%S').dt.hour  # Only extract the hour
traffic_by_hour = df_updated_data.groupby('hour')['traffic_level'].mean()
traffic_by_hour.head()
```

#### Visualization

```python
#Plotting the data using Line Plot
plt.figure(figsize=(10,6))
plt.plot(traffic_by_hour.index,traffic_by_hour.values, marker='o', color='g', linestyle='-', linewidth=2)
plt.title("Average Traffic Level Grouped by Time")
plt.xlabel("Time(Hour)")
plt.ylabel("Average Traffic Level")
plt.xticks(range(24), rotation=45) 
plt.grid(True)
plt.subplots_adjust()
plt.show()
```

```python
#Plotting the data using Bar Plot
plt.figure(figsize=(10, 6))
traffic_by_hour.plot(kind='bar', color='skyblue')
plt.title("Average Traffic Level by Time(Hour)")
plt.xlabel("Time(Hour)")
plt.ylabel("Average Traffic Level")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
